```python
# library load
import pandas as pd
import pyarrow
import numpy as np
import nflreadpy as nfl

from tabulate import tabulate
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
KEEP_COLS = [
    'game_id',
    'season',
    'week',
    'home_team',
    'away_team',
    'result',
    'play_type',
    'play_id',
    'posteam',
    'defteam',
    'rush_attempt', 
    'pass_attempt',
    'epa']

def load_schedule_pbp(season):
    df = nfl.load_pbp(season).to_pandas().assign(season=season)
    df = df[KEEP_COLS]
    return df

data = pd.concat([load_schedule_pbp(season) for season in range(1999, 2025)])
```

```python
data.shape
```

```python
plays = data[data['play_type'].notna() & data['play_type'].isin(['pass', 'run', 'punt', 'field_goal', 'kickoff', 'no_play'])]

plays = plays.sort_values(['season', 'week', 'game_id', 'play_id'])
plays['play_id'] = plays.groupby('game_id').cumcount() + 1

print(f"Total rows: {len(data)}, Actual plays: {len(plays)}")
```

```python
plays.to_csv('../data/raw_pbp_data.csv', index=False)
```
