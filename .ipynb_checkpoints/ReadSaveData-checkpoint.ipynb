```python
# library load
import pandas as pd
import pyarrow
import numpy as np
import nflreadpy as nfl

from tabulate import tabulate
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pbp = nfl.load_pbp().to_pandas()
### code for reading base dataset
chi_b = nfl.load_team_stats(seasons=True).to_pandas()
chi_b = chi_b[chi_b['team']=='CHI'].copy()
chi_b = chi_b.reset_index(drop=True)
chi_b.head()
```

```python
chi_game_results = nfl.load_schedules().to_pandas()
chi_win_stats = chi_game_results[(chi_game_results['away_team']=="CHI") | (chi_game_results['home_team']=="CHI")]
chi_win_stats[['season','week','away_team','away_score','home_team','home_score','result']].head()
```

```python
keep_cols = ['season','week','away_team','away_score','home_team','home_score','result']
chi_win_stats = chi_win_stats[keep_cols]
chi_win_stats['chi_win'] = (
    ((chi_win_stats['home_team'] == 'CHI') & (chi_win_stats['result'] > 0)) |
    ((chi_win_stats['away_team'] == 'CHI') & (chi_win_stats['result'] < 0))
).astype(int)
chi_win_stats.head()
```

```python
#chi_b = chi_b.rename({'team':''})
chi_b.head()
```

```python
print(chi_b.shape)
print(chi_win_stats.shape)
```

```python
chi_full = pd.merge(chi_b, chi_win_stats, on=['season','week'], how='inner')
chi_full.shape
```
